```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  OSRNS
sectors = [
    "UI ", "", "  ", "BPMN",
    " .", "IT for IT", " ",
    " ", "", "", "", " /"
]
completion = [100, 100, 90, 50, 80, 20, 100, 100, 100, 90, 100, 100]

#      
angles = np.linspace(0, 2 * np.pi, len(sectors), endpoint=False)

#  ( ) -  
radii = completion

#    ( )
width = 2 * np.pi / len(sectors)

#     
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

#  -
bars = ax.bar(angles, radii, width=width, bottom=0, color='blue', edgecolor='black', alpha=0.8)

#      
for angle, radius, sector in zip(angles, radii, sectors):
    rotation = np.degrees(angle)
    alignment = 'left'
    if rotation > 90 and rotation < 270:
        rotation += 180
        alignment = 'right'
    ax.text(angle, radius + 20, sector, rotation=rotation, ha=alignment, va='center', fontsize=10)

#        
for angle, radius in zip(angles, radii):
    ax.text(angle, radius - 5, f"{radius}%", ha='center', va='center', fontsize=9, color='white', fontweight='bold')

#   
ax.set_ylim(0, 130)

#     
ax.set_yticklabels([])
ax.set_xticklabels([])

plt.title(' ', fontsize=14, pad=20)
#    
plt.savefig('diagram.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#  
sectors = [
    "UI ", "", " ", "",
    "BPMN", "UI ", " ",
    "  ", " ",
    "  1 ", "  "
]
completion = [100, 100, 0, 100, 50, 70, 80, 0, 0, 0, 0]

#    
angles = np.linspace(0, 2 * np.pi, len(sectors), endpoint=False)
width = 2 * np.pi / len(sectors)

# 
fig, ax = plt.subplots(figsize=(9, 9), subplot_kw={'projection': 'polar'})
bars = ax.bar(angles, completion, width=width, bottom=0, color='mediumpurple', edgecolor='black', alpha=0.8)

#     
for angle, radius, sector in zip(angles, completion, sectors):
    rotation = np.degrees(angle)
    ha = 'left' if 0 <= rotation <= 180 else 'right'
    if 90 < rotation < 270:
        rotation += 180
    ax.text(angle, 120, sector, ha=ha, va='center', rotation=rotation, fontsize=9)

#     
for angle, radius in zip(angles, radii):
    ax.text(angle, radius - 5, f"{radius}%", ha='center', va='center', fontsize=9, color='white', fontweight='bold')

#   
ax.set_ylim(0, 130)

#     
ax.set_yticklabels([])
ax.set_xticklabels([])

plt.title('', fontsize=14, pad=20)
#    
plt.savefig('diagram.png', dpi=300, bbox_inches='tight')
plt.show()

```
