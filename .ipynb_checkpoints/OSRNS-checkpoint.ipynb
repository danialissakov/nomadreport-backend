```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Данные OSRNS
sectors = [
    "UI Модуля", "Сервисы", "Калькуляторы в АИС", "BPMN",
    "Выпуск доп.соглашений", "IT for IT", "Печатные формы",
    "Выпуск договора", "Транши", "Филиалы", "Автоандеррайтинг", "Выпуск ЭСФ/АВР"
]
completion = [100, 100, 90, 50, 80, 20, 100, 100, 100, 90, 100, 100]

# Преобразуем данные в радианы для углов
angles = np.linspace(0, 2 * np.pi, len(sectors), endpoint=False)

# Радиус (высота лепестков) - процент выполнения
radii = completion

# Ширина каждого лепестка (равномерно распределены)
width = 2 * np.pi / len(sectors)

# Создаем график в полярных координатах
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# Строим столбцы-лепестки
bars = ax.bar(angles, radii, width=width, bottom=0, color='blue', edgecolor='black', alpha=0.8)

# Добавим подписи с названиями снаружи лепестков
for angle, radius, sector in zip(angles, radii, sectors):
    rotation = np.degrees(angle)
    alignment = 'left'
    if rotation > 90 and rotation < 270:
        rotation += 180
        alignment = 'right'
    ax.text(angle, radius + 20, sector, rotation=rotation, ha=alignment, va='center', fontsize=10)

# Добавим подписи с процентами прямо на вершинах лепестков
for angle, radius in zip(angles, radii):
    ax.text(angle, radius - 5, f"{radius}%", ha='center', va='center', fontsize=9, color='white', fontweight='bold')

# Настроим лимиты радиуса
ax.set_ylim(0, 130)

# Уберем оси радиуса и углов
ax.set_yticklabels([])
ax.set_xticklabels([])

plt.title('Модуль ОСРНС', fontsize=14, pad=20)
# Сохранить картинку в файл
plt.savefig('diagram.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Данные ДПА
sectors = [
    "UI Модуля", "Сервисы", "ВНД дивиденды", "АИС",
    "BPMN", "UI Дивиденды", "Печатки дивиденды",
    "Печатка доля участия", "Печатка удержание",
    "Интеграция с 1С дивиденды", "Модуль выплат дивиденды"
]
completion = [100, 100, 0, 100, 50, 70, 80, 0, 0, 0, 0]

# Углы и ширина лепестков
angles = np.linspace(0, 2 * np.pi, len(sectors), endpoint=False)
width = 2 * np.pi / len(sectors)

# График
fig, ax = plt.subplots(figsize=(9, 9), subplot_kw={'projection': 'polar'})
bars = ax.bar(angles, completion, width=width, bottom=0, color='mediumpurple', edgecolor='black', alpha=0.8)

# Названия секторов — вне круга
for angle, radius, sector in zip(angles, completion, sectors):
    rotation = np.degrees(angle)
    ha = 'left' if 0 <= rotation <= 180 else 'right'
    if 90 < rotation < 270:
        rotation += 180
    ax.text(angle, 120, sector, ha=ha, va='center', rotation=rotation, fontsize=9)

# Проценты — на вершине лепестка
for angle, radius in zip(angles, radii):
    ax.text(angle, radius - 5, f"{radius}%", ha='center', va='center', fontsize=9, color='white', fontweight='bold')

# Настроим лимиты радиуса
ax.set_ylim(0, 130)

# Уберем оси радиуса и углов
ax.set_yticklabels([])
ax.set_xticklabels([])

plt.title('ДПА', fontsize=14, pad=20)
# Сохранить картинку в файл
plt.savefig('diagram.png', dpi=300, bbox_inches='tight')
plt.show()

```
